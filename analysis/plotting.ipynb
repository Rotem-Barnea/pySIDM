```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

data = pd.read_csv('baryonic all positions.csv')
fig,ax = plt.subplots(figsize=(6,5))
sns.kdeplot(data,log_scale=True,ax=ax)
ax.grid()
ax.grid(True, which='minor', alpha=0.3, linestyle='--')
ax.set_xlabel('Radius [kpc]')
ax.set_ylabel(r'Particle density per unit radius ($\propto \rho\cdot r^2$)')
ax.set_title('Baryonic matter radius distribution')
```

```python
import pandas as pd
import scipy
from matplotlib import pyplot as plt
import seaborn as sns

binned_data = pd.read_csv('baryonic binned data.csv')
fig,ax = plt.subplots(figsize=(6,5))
for column in ['initial','max core','collapse']:
    sub = binned_data[binned_data[f'{column} density'] > 0]
    x = sub['bin_centers']
    y = scipy.ndimage.gaussian_filter1d(sub[f'{column} density'], sigma=1)
    sns.lineplot(x=x,y=y,ax=ax,label=column)
ax.grid()
ax.grid(True, which='minor', alpha=0.3, linestyle='--')
ax.set(xscale='log',yscale='log')
ax.set_xlabel('Radius [kpc]')
ax.set_ylabel(r'$\rho\ \left[\frac{M_\odot}{kpc^3}\right]$')
ax.set_title('Baryonic matter radius distribution')
ax.legend()
```
